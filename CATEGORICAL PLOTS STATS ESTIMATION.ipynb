```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

## Countplot()

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.countplot(x = 'division',data = df)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df)
```

## Adding another Parameter Hue

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.countplot(x = 'level of education',data = df,hue = 'division')
```

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.countplot(x = 'level of education',data = df,hue = 'division',palette = 'Dark2')
```

## Barplot()

```python
plt.figure(figsize = (10,4),dpi=100)
sns.barplot(x = 'level of education',y = 'salary',data = df,estimator=np.mean,ci = 'sd')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')
```

```python
plt.figure(figsize=(12,6),dpi=100)

sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.histplot(x = 'salary',data = df)
```

```python
plt.figure(figsize = (12,6),dpi = 100)
sns.heatmap(df.corr())
```
